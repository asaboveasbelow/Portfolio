#### We can see the length of time the empoloyees spent in their workplace. We extract another column with the information.

#### To discover a reason of changing job, we need to dive into columns and extract some pattern from responses, which were given by survey.

```python
tafe_resignations['contributing_factors._dissatisfaction'].value_counts()
```

```python
tafe_resignations['contributing_factors._job_dissatisfaction'].value_counts()
```

```python
def update_vals(x):
    if x == '-':
        return False
    elif pd.isnull(x):
        return np.nan
    else:
        return True
tafe_resignations['dissatisfied'] = tafe_resignations[['contributing_factors._dissatisfaction', 'contributing_factors._job_dissatisfaction']].applymap(update_vals).any(1, skipna=False)
tafe_resignations_up = tafe_resignations.copy()
```

#### From the results easy to notice that 91 people were dissatisfied from their job.

```python
tafe_resignations_up['dissatisfied'].value_counts(dropna=False)
```

#### From the results easy to notice that 149 people were dissatisfied from their job.

```python
dete_resignations['dissatisfied'] = dete_resignations[['job_dissatisfaction',
       'dissatisfaction_with_the_department', 'physical_work_environment',
       'lack_of_recognition', 'lack_of_job_security', 'work_location',
       'employment_conditions', 'work_life_balance',
       'workload']].any(1, skipna=False)
dete_resignations_up = dete_resignations.copy()
dete_resignations_up['dissatisfied'].value_counts(dropna=False)
```

#### By adding a column, we can differented date from each group.

```python
dete_resignations_up['institute'] = 'DETE'
tafe_resignations_up['institute'] = 'TAFE'
```

#### Join these two group to compare values, and remove unnecessary columns.

```python
combined = pd.concat([dete_resignations_up, tafe_resignations_up], ignore_index=True)
```

```python
combined.notnull().sum().sort_values()
```

#### Removing columns where are less than 500 non null values to make more clear date.

```python
combined_updated = combined.dropna(thresh = 500, axis =1).copy()
```

```python
combined_updated.info()
```

#### It matters career stage, according to article from Businesswire (https://www.businesswire.com/news/home/20171108006002/en/Age-Number-Engage-Employees-Career-Stage)

```python
combined_updated['institute_service_up'] = combined_updated['institute_service'].astype('str').str.extract(r'(\d+)')
combined_updated['institute_service_up'] = combined_updated['institute_service_up'].astype('float')
```

```python
combined_updated['institute_service_up'].value_counts()
```

#### We need to categorize how long their work.

```python
def transform_service(val):
    if val >= 11:
        return "Veteran"
    elif 7 <= val < 11:
        return "Established"
    elif 3 <= val < 7:
        return "Experienced"
    elif pd.isnull(val):
        return np.nan
    else:
        return "New"
combined_updated['service_cat'] = combined_updated['institute_service_up'].apply(transform_service)
```

```python
combined_updated['service_cat'].value_counts()
```

#### We still have to cope with missing values.

```python
combined_updated['dissatisfied'].value_counts(dropna=False)
```

#### Repalcing missing values with most common value.

```python
combined_updated['dissatisfied'] = combined_updated['dissatisfied'].fillna(False)
```

#### To compare each group where people were disconnected with work.

```python
dis_pct = combined_updated.pivot_table(index='service_cat', values='dissatisfied')
```

```python
%matplotlib inline
dis_pct.plot(kind='bar', rot=30)
```

#### From the initial analysis above, we can tentatively conclude that employees with 7 or more years of service are more likely to resign due to some kind of dissatisfaction with the job than employees with less than 7 years of service. However, we need to handle the rest of the missing data to finalize our analysis.
